 ## Preprocessing and training

```python
# Train/Test split. We do have some missing values for some features and we are going to use sklearn's pipelines to scale and impute the data.
data2.head()
```

```python
from sklearn.model_selection import train_test_split, cross_validate, GridSearchCV, learning_curve
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor 
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.feature_selection import SelectKBest, f_regression
```

```python
#train/test split
X_train,X_test,y_train,y_test = train_test_split(data2.drop(columns = 'Life_expectancy_'),data2.Life_expectancy_ , test_size = 0.3, random_state = 47)
```

```python
X_train.shape,X_test.shape
```

```python
y_train.shape, y_test.shape
```

```python
names_train = X_train[['Country','Year','Status']]
names_test = X_test[['Country','Year','Status']]
X_train.drop(columns=['Country','Year','Status'], inplace=True)
X_test.drop(columns=['Country','Year','Status'], inplace=True)
X_train.shape, X_test.shape
```

```python
X_train.dtypes
```

```python
X_test.dtypes
```

```python
X_train.head()
```

```python
pipe_LR = make_pipeline(SimpleImputer(strategy = 'median'), StandardScaler(),SelectKBest(f_regression),LinearRegression())
```

```python
cv_results = cross_validate(pipe_LR, X_train, y_train, cv = 5)
```

```python
cv_scores = cv_results['test_score']
cv_scores
```

```python
np.mean(cv_scores), np.std(cv_scores)
```

```python
pipe_LR.get_params().keys()
```

```python
k= [k+1 for k in range(len(X_train.columns))]
grid_params ={'selectkbest__k':k}
```

```python
grid_LR = GridSearchCV(pipe_LR, param_grid = grid_params, cv = 5, n_jobs=-1)
```

```python
grid_LR.fit(X_train,y_train)
```

```python
grid_LR.best_params_
```

```python
grid_LR.best_score_
```

```python
grid_LR.best_estimator_
```

```python
score_mean = grid_LR.cv_results_['mean_test_score']
score_std = grid_LR.cv_results_['std_test_score']
cv_k = [k for k in grid_LR.cv_results_['param_selectkbest__k']]
```

```python
best_k = grid_LR.best_params_['selectkbest__k']
plt.subplots(figsize=(10, 5))
plt.errorbar(cv_k, score_mean, yerr=score_std)
plt.axvline(x=best_k, c='r', ls='--', alpha=.5)
plt.xlabel('k')
plt.ylabel('CV score (r-squared)')
plt.title('Pipeline mean CV score (error bars +/- 1sd)');
```

```python
selected = grid_LR.best_estimator_.named_steps.selectkbest.get_support()
selected
```

```python
model_coeficients = grid_LR.best_estimator_.named_steps.linearregression.coef_
features = X_train.columns[selected]
pd.Series(model_coeficients, index = features).sort_values(ascending= False)
```

#### Random Forest Regression model

```python
cv_scores_RF = cv_results_RF['test_score']
cv_scores_RF
```
